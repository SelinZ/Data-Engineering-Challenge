```python
# Detect all NaN, None, or empty cells
nan_or_empty = df1.isnull() | df1.applymap(lambda x: str(x).strip() == '')
print("Cells that are NaN, None, or empty:")
print(nan_or_empty)
```

```python
columns_with_issues = nan_or_empty.any(axis=0)
print("\nColumns with NaN, None, or empty cells:")
print(columns_with_issues[columns_with_issues].index)
```
